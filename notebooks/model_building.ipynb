## Part 4: Build a regression model.

```python
# Import the required libraries
import pandas as pd
import numpy as np

# statsmodels modeling library
import statsmodels.api as sm

# plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

# suppress minor warnings from seaborn
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in the yelp data from part 2 and 3.
yelp_df = pd.read_csv("../data/yelp_poi_data.csv")
yelp_df
```

```python
# Assign the appropriate dataframe columns to the dependent and independent variables of the OLS model. 
# In this analysis, we are looking at the relationship between the distance from the bike station to the point of interest and the POI's yelp rating

X = yelp_df['yelp_distance']
y = yelp_df['yelp_rating']
```

Provide model output and an interpretation of the results. 

```python
X = sm.add_constant(X) # adding a constant
lin_reg = sm.OLS(y,X)
```

```python
model = lin_reg.fit()
print_model = model.summary()
print(print_model)
```

Above, we see an extremely weak relationship between distance and rating. The relationship is statistically significant at over p = 0.05.
